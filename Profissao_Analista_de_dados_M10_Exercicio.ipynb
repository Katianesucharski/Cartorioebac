## 1\. Excel para CSV

Utilizando o pacote Python `openpyxl` visto em aula, extraia os seguintes as colunas `id`, `sexo` e `idade` para dos clientes inadimplentes (`default = 1`) e solteiros (`estado_civil = 'solteiro'`). Salves os dados extraídos no arquivo csv `credito.csv` separado por `;`. Exemplo do cabeçalho e das três primeiras linhas:

```
id;sexo;idade
767712558;59;M
713741358;46;M
772390908;59;M
```

**Dica:** O arquivo csv `credito.csv` deve ter 669 linhas, contando com o cabeçalho.

**Nota:** Escreva o código da sua solução abaixo em uma ou mais células, você não precisa enviar o arquivo csv gerado.

````
from  openpyxl import load_workbook

planilhas = load_workbook(filename='credito.xlsx')
planilha = planilhas.active
````

````
clientes_inadiplentes = []

for row in planilha.iter_rows(min_row=2):
  cliente_id = row[0].value
  sexo = row[2].value
  idade = row[3].value
  default = row[1].value

  if default == 1:
    clientes_inadiplentes.append([cliente_id,sexo,idade])

clientes_solteiros = []
for solteiros in clientes_inadiplentes:
    estado_civil  = solteiros
    if estado_civil == 'solteiro':
        clientes_solteiros.append(solteiro)




````

````
import csv


cabecalho = ["Cliente ID", "Idade", "Sexo"]


nome_arquivo = "credito.csv"


with open(file=nome_arquivo, mode="w", encoding="utf8", newline="") as arquivo:
    cswriter = csv.writer(arquivo, delimiter=",")
    cswriter.writerow(cabecalho)


with open(file=nome_arquivo, mode="a", encoding="utf8", newline="") as arquivo:
    cswriter = csv.writer(arquivo, delimiter=",")
    for cliente in clientes_inadiplentes:
        cswriter.writerow(cliente)

print("Dados dos clientes inadimplentes foram escritos no arquivo", nome_arquivo)



````

Extraia os números de contato do arquivo texto `ebac.txt` e salve-os no arquivo csv `contato_ebac.csv` com o separador `;` no seguinte formato:

```
tipo;numero
whatsapp;+551142002991
```

**Nota:** Escreva o código da sua solução abaixo em uma ou mais células, você não precisa enviar o arquivo csv gerado.

````
with open(file='./ebac.txt', mode='r', encoding='utf8') as arquivo:
  linhas = arquivo.readlines()

print(linhas)
````

````
linhas = filter(lambda linha: linha != '\n', linhas)
linhas = map(lambda linha: linha.strip(), linhas)
linhas = list(linhas)

print(linhas)
````

````
linhas_com_contato = filter(lambda linha: '+55' in linha, linhas)
linhas_com_contato = list(linhas_com_contato)

print(linhas_com_contato)
````
